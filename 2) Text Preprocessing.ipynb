**Terminologies used in NLP**

- Corpus : we get a Paragraph which is called corpus
- Documents : Sentences in a paragraph are called as documents
- Vocabulary : **Unique words** in a corpus, example dictionary
- Words : Words in a corpus

There are 2 important libraries used in NLP 
- NLTK
- Spacy

Key Differences
- Performance: SpaCy is significantly faster and more memory-efficient than NLTK, especially for large-scale processing.
- Ease of Use: SpaCy has a simpler API and handles many tasks automatically, while NLTK requires more code and configuration.
- Use Cases: Use NLTK for academic research, prototyping, or when you need fine-grained control over algorithms. Use SpaCy for production apps, chatbots, or when speed and accuracy are priorities.

**Tokenisation**

Tokenisation is the process in which we take the sentences or paragraphs aka corpus and break down into smaller units called tokens. 

Importantly tokenisation could also be breaking down paragraphs into sentences or sentences to words. It is not just limited to breaking down to the smallest form. 

for a Corpus : "My name is Hemant and I like to learn data science topics like ML, DL, NLP. I am currently employed with an MNC."

Tokenisation is 
- My name is Hemant and I like to learn data science topics like ML, DL, NLP
- I am currently employed with an MNC

**Stemming**

Stemming is the process of reducing a word to its word stem or to the root of words known as lemma. 
- eating, eaten, eat : All can be condensed down to be eat (root verb : eat). This increases the number of input features also each of these when needed ot be converted to vectors can be expensive, it will be like having similar kind of words
