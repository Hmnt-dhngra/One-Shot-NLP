```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
words =["eating","eaten","eats","writes","programs", "programming", "history", "congratulations"]
stemmed_words = [stemming.stem(word) for word in words]
print("Stemmed words:", stemmed_words)
```

**RegexpStemmer Class**

Using this class we can easily implement Regular Expression Stemmer algortihms. It bascially takes a single regular expression and removes any prefix or suffix that matches the expressions. 

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)## min = minimum length of string to stem
```

```python
reg_stemmer.stem("programming")
```

```python
words =["eating","eaten","eats","writes","programs", "programming", "history", "congratulations"]
stemmed_words_regex = [reg_stemmer.stem(word) for word in words]
print("Stemmed words:", stemmed_words_regex)
```

```python
reg_stemmer.stem("ingeating")
```

```python
reg_stemmer.stem("inning")
```

```python
reg_stemmer.stem("lingsu")
```

**Snowballstemmer** 

It is an improvement over the Porter Stemmer. It is more aggressive and can handle a wider range of words. It is also faster than the Porter Stemmer.

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer(language="english")
```

```python
words =["eating","eaten","eats","writes","programs", "programming", "history", "congratulations"]
stemmed_words_snowball = [snowball_stemmer.stem(word) for word in words]
print("Stemmed words:", stemmed_words_snowball)
```

```python
stemming.stem('fairly'), stemming.stem('sportingly')
```

```python
snowball_stemmer.stem('fairly'), snowball_stemmer.stem('sportingly') ## Snowball Stemmer is more aggressive than Porter Stemmer, it removes the 'ly' suffix from both words.
```

```python
snowball_stemmer.stem('goes')
```

```python
stemming.stem('goes')
```
