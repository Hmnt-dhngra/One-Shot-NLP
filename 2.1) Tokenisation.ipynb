```python
corpus = """
Artificial Intelligence (AI's) is don't revolutionizing industries by automating tasks that once required human intelligence—think chatbots, self-driving cars, and predictive analytics!
However, challenges like bias in algorithms and ethical concerns (e.g., privacy issues with data mining) must be addressed to ensure responsible development. 
With advancements in machine learning, AI systems can now process complex data streams, including text with emojis 😀, hashtags #AI, and symbols like @mentions, making interactions more dynamic and engaging.

"""
```

```python
from nltk.tokenize import sent_tokenize # to convert paragraph into sentences
```

```python
documents=sent_tokenize(corpus)
type(documents)
```

```python
for doc in documents:
    print(doc)
```

```python
from nltk.tokenize import word_tokenize # to convert paragraph into words or sentences into words
```

```python
word_tokenize(corpus)
```

```python
for doc in documents:
    print(word_tokenize(doc))
```

```python
from nltk.tokenize import wordpunct_tokenize # to convert paragraph into words or sentences into words but also considers punctuation as separate tokens
```

```python
for doc in documents:
    print(wordpunct_tokenize(doc)) 
```

notice the splitting of AI's in this cell and the cell above. In the cell above, it is considered as one token but in this cell, it is split into two tokens: AI and 's. This is because wordpunct_tokenize considers punctuation as separate tokens, while word_tokenize does not.

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
out=tokenizer.tokenize(corpus)
print(out) 
```

check the word development...The full stop is part of the tokenised word here development. This is because TreebankWordTokenizer considers punctuation as part of the token if it is attached to a word, while word_tokenize and wordpunct_tokenize do not. But for the last full stop, it will be separate word.
